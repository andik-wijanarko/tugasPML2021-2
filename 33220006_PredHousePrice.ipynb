```python
import tensorflow as tf
from tensorflow.keras.datasets import boston_housing
from sklearn import preprocessing
```

memuat datasets boston house price ke dalam parameter

```python
(train_x,train_y),(test_x,test_y)=boston_housing.load_data()
```

normalisasi value agar value berisi angka antara 0..1

```python
train_x=preprocessing.normalize(train_x)
test_x=preprocessing.normalize(test_x)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
```

#Membuat model, 
model terdiri dari multilayer preceptron dengan 4 layer.
layer-1 terdiri dari 128 neuron. untuk fitur input dari dataset. 
layer-2 hidden layer dengan 64 neuron.
layer-3 hidden layer dengan 32 neuron.
layer-4 adalah layer output dengan 1 neuron untukhasil prediksi harga rumah.


```python
def HousePricePredictionModel():
  model=Sequential()
  model.add(Dense(128,activation='relu',input_shape=(train_x[0].shape)))
  model.add(Dense(64,activation='relu'))
  model.add(Dense(1))
  model.compile(optimizer='Adam',loss='mse', metrics=['mae'])
  return model
```

persiapan training dengan epocs 100

```python
import numpy as np
k=4
num_val_samples=len(train_x)
num_epochs=100
all_scores=[]
```

training dataset 

```python
model=HousePricePredictionModel()
model.fit(x=train_x,y=train_y,epochs=num_epochs,batch_size=1,verbose=1,validation_data=(test_x,test_y))
```
